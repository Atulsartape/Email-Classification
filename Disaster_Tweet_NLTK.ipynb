```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/tweets.csv')
df.head()
```

```python
X = df['text']
y = df['target']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,stratify=y)
```

```python
import gensim
```

```python
from gensim.models import word2vec
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
import nltk
nltk.download('punkt_tab')
```

```python
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('english'))  # Use "english" stopwords
print(stop_words)  # Displays the list of stopwords
```

```python
def remove_urls(text):

  text = re.sub(r'http\S+|www\S+|https\S+|t\.co\S+', '', text, flags=re.MULTILINE)
  return text

data = []
for text in X_train:
  text = remove_urls(text)
  for sentence in sent_tokenize(text):
    temp = []
    for word in word_tokenize(sentence):
      clean_word = re.sub('[^a-z]', '', word)
      clean_word1 = re.sub(r'(http\S+|www\S+|https\S+|\S+\.\S+)', '', clean_word)
      if clean_word1 and clean_word1.lower() not in stop_words:

        temp.append(word)
    data.append(temp)


```

```python
len(X_test)
```

```python
print(data)
```

```python
from gensim.models import Word2Vec
model1 = Word2Vec(data, min_count=1, vector_size=100, window=5)
```

```python
def vectorize_sentence(sentence, model):
    words = word_tokenize(sentence)
    word_vectors = [model.wv[word] for word in words if word in model.wv]
    if len(word_vectors) == 0:
        return np.zeros(model.vector_size)
    return np.mean(word_vectors, axis=0)

X_train_data = np.array([vectorize_sentence(sentence, model1) for sentence in X_train])
X_test_data = np.array([vectorize_sentence(sentence, model1) for sentence in X_test])
```

```python
X_train_data
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
clf = LogisticRegression()
```

```python
clf.fit(X_train_data,y_train)
```

```python
y_pred = clf.predict(X_test_data)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
